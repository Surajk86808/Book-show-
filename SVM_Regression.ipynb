```python
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df[['total_bill']]
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.head()
```

```python
#feature encoding and label onehot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
```

```python
lb1 = LabelEncoder()
lb2 = LabelEncoder()
lb3 = LabelEncoder()
# why three label encoder?
# beacuse there are three binary classification answer time,smoker,sex
```

```python
X_train ['sex']= lb1.fit_transform(X_train['sex'])
X_train ['smoker']= lb2.fit_transform(X_train['smoker'])
X_train ['time']= lb3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

label encoder changes the data to binary zero and ones

```python
X_test ['sex']= lb1.fit_transform(X_test['sex'])
X_test ['smoker']= lb2.fit_transform(X_test['smoker'])
X_test ['time']= lb3.fit_transform(X_test['time'])
```

```python
X_test.head()
```

```python
# One hot encoding columns transfer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('onehot', OneHotEncoder(drop='first'), ['day'])], remainder='passthrough')
```

```python
X_train_transformed = ct.fit_transform(X_train)
```

```python
X_test_transformed = ct.transform(X_test)
```

```python
from sklearn.svm import SVR, SVC
model = SVR()
```

```python
model.fit(X_train_transformed,y_train)
```

```python
y_pred = model.predict(X_test_transformed)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```
